```python
import hoomd
import gsd.hoomd
import numpy as np
import sys 
import matplotlib.pyplot as plt


sys.path.append("..")
from src import smd
from src.utilities import ChIMES

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Helvetica'] + plt.rcParams['font.sans-serif']
plt.rcParams['mathtext.it'] = 'Helvetica:italic'

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize']=(2, 2)
plt.rcParams['font.size'] = 8
plt.rcParams['axes.titlesize'] = 8
plt.rcParams['axes.labelsize'] = 8
plt.rcParams['axes.titlesize'] = 8
plt.rcParams['figure.titlesize'] = 8
plt.rcParams['legend.fontsize'] = 8
plt.rcParams['lines.linewidth'] = 1
plt.rcParams['lines.markersize'] = 3.0
```

```python
def LJ(r, epsilon, sigma):
    quotient = sigma / r 
    quotient_6 = quotient * quotient * quotient * quotient * quotient * quotient
    
    return 4 * epsilon * (quotient_6 * quotient_6 - quotient_6)
```

```python
TIME_CONVERSION = 0.01  # From 1 fs to 100fs (derived time unit)
VELOCITY_CONVERSION = 1 / 0.01  # From 1 A/fs to 0.01 A/fs (=1 nm/ps)
```

```python
# particle information
dt = 1   # 1 = 0.1 ps = 100 fs
mass = 56.0
box_length = 100.0
init_p1 = np.array([0., 0., 0.])
init_p2 = np.array([3.0, 0., 0.])

spring_constant_pull = 50

ang_vel_conversion = 100  # convert ang vel to radian / timestep 

# thermo parameters
T = 10             # K
kb = 0.00831446262  # kJ/mol/K
kT = kb*T
tau = 10.0 * dt
tau *= TIME_CONVERSION

epsilon = 1.0
sigma = 1.0
r_cut = 6.0

# sim params
traj_fn = "traj.gsd"
dump_period = 1000
total_steps = 200_000
eq_steps = 50_000
seed = 1000
segment = 10_000
pull_velocity = -1.0E-5
pull_direction = np.array([1.0, 0.0, 0.0])
dt *= TIME_CONVERSION
pull_velocity *= VELOCITY_CONVERSION
```

```python
snapshot = gsd.hoomd.Frame()
snapshot.particles.N = 2
snapshot.particles.position = np.vstack((init_p1, init_p2))
snapshot.particles.types = ['A'] 
snapshot.particles.typeid = [0] * 2 
snapshot.configuration.box = [box_length, box_length, box_length, 0.0, 0.0, 0.0]
snapshot.particles.mass = [mass] * 2
snapshot.configuration.step = 0
```

```python
device = hoomd.device.auto_select()
sim = hoomd.Simulation(device=device, seed=seed)
sim.timestep = 0
sim.create_state_from_snapshot(snapshot)
```

```python
forces = []
nl = hoomd.md.nlist.Cell(buffer=0.5, default_r_cut=r_cut)  

lj = hoomd.md.pair.LJ(nlist=nl, default_r_cut=r_cut)
lj.params[('A', 'A')] = dict(epsilon=epsilon, sigma=sigma)
lj.r_cut[('A', 'A')] = r_cut

forces.append(lj)
```

```python
smd_force = smd.SMD_constV_couple(
        sim=sim,
        group1_idx=[0],
        group2_idx=[1],
        dt=dt,
        pulling_velocity=pull_velocity,
        spring_constant=spring_constant_pull,
        pulling_direction=pull_direction
    )
forces.append(smd_force)
```

```python
ensemble = hoomd.md.methods.ConstantVolume(
    filter=hoomd.filter.All(),
    thermostat=hoomd.md.methods.thermostats.MTTK(kT=kT, tau=tau)
)
```

```python

integrator = hoomd.md.Integrator(
    dt=dt,
    methods=[ensemble],
    forces=forces,
    integrate_rotational_dof=False
)
sim.operations += integrator
```

```python
logger = hoomd.logging.Logger()
logger.add(obj=lj, quantities=["energy"])

gsd_writer = hoomd.write.GSD(
    filename=traj_fn,
    trigger=hoomd.trigger.Periodic(dump_period),
    dynamic=["property"],
    mode='wb',
    logger=logger
)
sim.operations += gsd_writer
```

```python
sim.operations += smd_force.get_writer("smd.txt", dump_period)

sim.state.thermalize_particle_momenta(filter=hoomd.filter.All(), kT=kT)
sim.operations += hoomd.md.update.ZeroMomentum(hoomd.trigger.On(0))
```

```python
sim.run(0)
sim.operations.writers[-1].write()  

while sim.timestep < total_steps:
    sim.run(segment)
    for writer in sim.operations.writers:
        if hasattr(writer, 'flush'):
            writer.flush()

velocity = np.copy(smd_force._velocity)
smd_force._velocity = 0.
sim.run(eq_steps)
for writer in sim.operations.writers:
    if hasattr(writer, 'flush'):
        writer.flush()
    
smd_force._velocity = -velocity

while sim.timestep < (total_steps*2 + eq_steps):
    sim.run(segment)
    for writer in sim.operations.writers:
        if hasattr(writer, 'flush'):
            writer.flush()
```

```python
fn = np.genfromtxt("smd.txt", skip_header=1)
timestep = fn[:, 0]
work = fn[:, -1] 
distance = fn[:, -3]

num_frames = int(total_steps / dump_period) + 1
work_fw = work[:num_frames]
work_rv = work[-num_frames:][::-1]
distance = distance[:num_frames]

pmf = (work_fw - work_rv) / 2
pmf -= pmf[0]

plt.plot(distance, pmf, label="$W$")

lj = LJ(distance, epsilon=epsilon, sigma=sigma)
plt.plot(distance, lj, label="LJ potential")
plt.plot(distance, work_fw, "--", label=r"$w_F$")
plt.plot(distance, work_rv, "--", label=r"$w_R$")
plt.legend()

```

### For interpolation purpose, fit ChIMES model to generate 2B PMF for the future 3B PMF calculation.

```python
chimes = ChIMES()
```

```python
O2b = 18
chimes_rout = 3.0
tersoff_fo = 0.75
chimes_rin = 0.95
morse_lambda = 2 ** (1/6)
```

```python
x, xin, xout = chimes.morse_trans(distance, chimes_rin, chimes_rout, morse_lambda)
s = chimes.rescale_into_s(x, xin, xout)
```

```python
smooth = chimes.tersoff_smooth(distance, tersoff_fo, chimes_rout)
```

```python
A = chimes.make_Amatrix(s, O2b, smooth, 0)
```

```python
c = chimes.solve_LSQ_SVD(A, pmf, svd_regularization_ratio=1e-5)
pmf_fit = A @ c
```

```python
fig, ax = plt.subplots(1, 2, sharey=True, figsize=(3, 2))
ax[0].scatter(pmf-pmf[0], pmf_fit, color="k", alpha=0.2)
ax[0].plot([-1, 1], [-1, 1], color="r", transform=ax[0].transAxes, label="x=y")
ax[0].set_xlabel(r"$W_\mathrm{2B}$")
ax[0].set_ylabel(r"$W_\mathrm{2B, ChIMES}$")
ax[0].legend()
ax[1].plot(distance, pmf-pmf[0], label=r"$W_\mathrm{2B}$")
ax[1].plot(distance, pmf_fit, "--", label=r"$W_\mathrm{2B, ChIMES}$")
ax[1].set_xlabel("r")
ax[1].legend(bbox_to_anchor=(1.0, 1.04))
print(f"RMSE: {np.mean((pmf-pmf_fit)**2.0)}")
```
